```python
# Import the necessary libraries:
# - pandas for data handling
# - Matplotlib for plotting graphs
# - LinearRegression from sklearn for creating the model
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset from the CSV file
# The file contains year-wise per capita income data
file_path = "canada_per_capita_income.csv"
data = pd.read_csv(file_path)
```

```python
# Display the first 5 rows of the dataset
# This helps to understand the structure of the data
data.head(5)
```

```python
# Create a scatter plot to visualize the relationship between years and per capita income
%matplotlib inline
plt.xlabel("Years")
plt.ylabel("per capita income (US$)")
plt.scatter(data["year"], data["per capita income (US$)"], color='black')
```

```python
""" -------- Linear Regression --------

Linear Equation: y = mx + b

Where:
- m = Slope (also called Gradient)
- b = Y-intercept (value of y when x = 0)
- x = Independent variable (input)
- y = Dependent variable (output or prediction)
"""
```

```python
# Train the Linear Regression model using the year as input and per capita income as output
reg = LinearRegression()
reg.fit(data[["year"]], data["per capita income (US$)"])
```

```python
# Create a CSV file with only the years, which will be used to predict the per capita income for those years.
```

```python
# Read the CSV file which contains only the years.
years_to_predict = pd.read_csv('years_to_predict.csv')

# Use the trained model to predict income for the given years.
predicted_income = reg.predict(years_to_predict)

# Show the predicted income values.
predicted_income
```

```python
# Create a new column in the DataFrame to store the predicted income
years_to_predict["Predicted_income"] = predicted_income

# Save the updated data (years and predicted income) back to the CSV file
years_to_predict.to_csv("years_to_predict.csv", index=False)
```

```python
# Show both previous income and predicted income by year
%matplotlib inline
plt.xlabel("Years", fontsize=10)
plt.ylabel("Per capita income (US$)", fontsize=10)
plt.scatter(data["year"], data["per capita income (US$)"], color='black', marker='*', label="Previous Data (Canada Per Capita Income)")
plt.plot(years_to_predict['year'], predicted_income, color='gray', label="Predicted Income (by Linear Regression)")
plt.legend()
```
